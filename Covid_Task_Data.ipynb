```
import imgaug as ia
import imageio
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
from imgaug import augmenters as iaaa
import numpy as np
import glob
import cv2
```

```
path = glob.glob('/content/drive/MyDrive/dataset/normal/*.*g')
cv_img = []
for img in path:
     n = cv.imread(img)
     im_rgb = cv.cvtColor(n, cv.COLOR_BGR2RGB)
     cv_img.append(im_rgb)
plt.imshow(cv_img[1])
print(len(cv_img))
```

```
seq = iaaa.Sequential([
     iaaa.Affine(rotate=(-15,15)),
    #iaaa.Fliplr((1.0))
 ], random_order=True)
```

```
images_different_sizes = cv_img
# # augment them as one batch
images_aug = seq.augment_images(cv_img)
plt.imshow(images_aug[1])
```

```
from PIL import Image
for ind, i in enumerate(images_aug):
    im = Image.fromarray(i)
    im.save("/content/drive/MyDrive/dataset/normal/aug2normal{}.jpg".format(str(ind)))
```
